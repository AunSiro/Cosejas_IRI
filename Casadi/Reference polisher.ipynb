```python
library = bibtexparser.parse_file('references.bib')
```

### Remove duplicated blocks:

```python
duplicated_blocks = []
for ii in library.failed_blocks:
    if isinstance(ii, bibtexparser.model.DuplicateBlockKeyBlock):
        print(ii.key)
        duplicated_blocks.append(ii)
```

```python
library.remove(duplicated_blocks)
```

```python
for ii in library.blocks:
    if isinstance(ii, bibtexparser.model.ImplicitComment) :
        print(ii)
        library.remove(ii)
```

```python
library.blocks.sort(key = lambda entry: entry.key)
```

```python
def url_title(title, url):
    link = r'\href{' + url + r'}{' + title + '}'
    return link
```

```python
def purge_str(text):
    return text.replace("'", "").replace("`", "").replace("{", "").replace("}", "").replace(r"\ "[0], "").replace("\n", "")
```

```python
def print_block(block):
    print(block.key)
    for ii in block.fields_dict:
        print('\t', ii, ' : ', purge_str(block.fields_dict[ii].value) )
```

```python
def add_link(block):
    try:
        auth = purge_str(block.fields_dict['author'].value)
    except:
        auth = ''
    title = block.fields_dict['title'].value
    title_clean = purge_str(title)
    print(auth, title_clean)
    url = input()
    if url == '':
        return
    else:
        url_field = bibtexparser.model.Field(key = 'url', value = url)
        block.set_field(url_field)
        block.fields_dict['title'].value = url_title(title, url)
        print('\n\n')
```

```python
jj = 1
for ii in library.blocks:
    field_keys = ii.fields_dict.keys()
    title = ii.fields_dict['title'].value
    if 'href' in title:
        continue
    islink = True
    if 'url' in ii.fields_dict:
        url = ii.fields_dict['url'].value
    elif 'URL' in ii.fields_dict:
        url = ii.fields_dict['URL'].value
    elif 'doi' in ii.fields_dict:
        url = r'https://doi.org/' + ii.fields_dict['doi'].value
    elif 'DOI' in ii.fields_dict:
        url = r'https://doi.org/' + ii.fields_dict['DOI'].value
    else:
        islink = False
    
    if islink:
        ii.fields_dict['title'].value = url_title(title, url)
    else:
        print_block(ii)
        add_link(ii)
        jj +=1
```

```python
bibtexparser.write_file('references.bib', library)
```
