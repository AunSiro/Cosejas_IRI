## Pendulum exercise
$$
\begin{split}\begin{array}{lc}
\begin{array}{l}
\text{minimize:} \\
x(\cdot) \in \mathbb{R}^2, \, u(\cdot) \in \mathbb{R}
\end{array}
\quad \displaystyle \int_{t=0}^{T}{cos(x_0) \, dt}
\\
\\
\text{subject to:} \\
\\
\begin{array}{ll}
\left\{
\begin{array}{l}
\dot{x}_0 = x_1 \\
\dot{x}_1 = u - sin(x_0) \\
-u_{max} \le u \le u_{max} , \quad
\end{array} \right. & \text{for} \, 0 \le t \le T \\
x_0(0)=0, \quad x_1(0)=0, x_0(T) = pi , x_1(T) = 0
\end{array}
\end{array}\end{split}
$$
with $T=10$.

siendo $$x_0 = \theta$$ $$x_1 = \theta'$$

```python
n_interp = 500
t_end = 50
ini_guess_t = np.linspace(0,t_end,n_interp)

ini_guess_q = np.ones_like(ini_guess_t)*np.pi
cut1 = 141
ini_guess_q[:cut1] = -np.sin(ini_guess_t[:cut1]*0.9)*ini_guess_t[:cut1]*0.14
cut2 = 250
ini_guess_q[cut1:cut2] = -np.sin(ini_guess_t[cut1:cut2]*0.43+(0.9-0.43)*ini_guess_t[cut1])*np.pi#*ini_guess_t[cut1:cut2]*0.14
cut3 = 300
ini_guess_q[:cut3] = smooth(ini_guess_q, 6)[:cut3]


ini_guess_v = np.gradient(ini_guess_q, ini_guess_t)
ini_guess_v = smooth(ini_guess_v, 10)


ini_guess_a = np.gradient(ini_guess_v, ini_guess_t)
ini_guess_a = smooth(ini_guess_a, 10)

ini_guess_x = np.concatenate((ini_guess_q[:, np.newaxis], ini_guess_v[:, np.newaxis]), 1)
ini_guess_x_d = np.concatenate((ini_guess_v[:, np.newaxis], ini_guess_a[:, np.newaxis]), 1)


ini_guess_u = np.zeros_like(ini_guess_t)
ini_guess_u[:230] = 0.2*np.sign(ini_guess_v)[:230]
```

```python
@lru_cache(maxsize=None)
def casadi_pendulum(
    scheme,
    N,
    ini_guess='lin',
    solve_repetitions = 1,
    t_end = 50,
    order = 2,
    u_m = 0.3,
    verbose = False,
    silent = True
):
    _opti = Opti_Problem(
        LM=F,
        params = params,
        scheme = scheme,
        ini_guess= ini_guess,
        t_end = t_end,
        verbose=verbose,
        silent=silent,
    )
    _opti.dynamic_setup('f_x', 2//order, 1, order = order)
    _opti.opti_setup(N)
    _opti.apply_scheme()
    
    opti = _opti.opti
    X = _opti.opti_arrs['x']
    X_d = _opti.opti_arrs['x_d']
    U = _opti.opti_arrs['u']
    T = _opti.opti_arrs['t']
    X_s = _opti.opti_points['x_s']
    X_e = _opti.opti_points['x_e']
    
    
    if ini_guess == 'lin':
        if order == 2:
            _opti.initial_guess([0,], [np.pi,])
        elif order == 1:
            _opti.initial_guess([0,np.pi/t_end], [np.pi,np.pi/t_end])
    if ini_guess == 'custom':
        _opti.opti.set_initial(
            X,
            interp_2d(T, ini_guess_t, ini_guess_x)
        )
        _opti.opti.set_initial(
            X_d,
            interp_2d(T, ini_guess_t, ini_guess_x_d)
        )
    _opti.u_sq_cost()
    
        
    #cost = cas.sum1(cas.cos(X[:,0]))    
    #_opti.cost = cost
    #_opti.opti.minimize(cost)
    
    
   # Initial and final conditions
    opti.subject_to(X_s.T == [0, 0])
    opti.subject_to(cas.cos(X[-1,0]) < -0.999)
    #opti.subject_to(T < t_m)
    opti.subject_to(opti.bounded(-0.001,X[-1,1],0.001))
    #u_init = np.interp(_opti.opti_arrs['t_col'], tt_in, uu_in)
    #opti.set_initial(U, u_init)
    
    # Appliying restrictions and action boundaries
    for ii in range(N):
        opti.subject_to(opti.bounded(-u_m,U[ii,:],u_m))
    if _opti.scheme_mode == 'equispaced':
        opti.subject_to(opti.bounded(-u_m,U[-1, :],u_m))
        if 'parab' in scheme:
            U_c = _opti.opti_arrs['u_c']
            for ii in range(N):
                #opti.set_initial(U_c, (u_init[:-1] + u_init[1:])/2)
                opti.subject_to(opti.bounded(-u_m,U_c[ii,:],u_m))
    
    _opti.chrono_solve(solve_repetitions=solve_repetitions,)
    return _opti.results
```
