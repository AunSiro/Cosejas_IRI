```python
from numpy import pi, arange, cos, array
def CG2(N):
    theta = pi * (2 * arange(N, dtype="float128") + 1) / N / 2
    tau = cos(theta)[::-1]
    tau = (tau -tau[::-1])/2
    tau = array(tau, dtype="float64")
    return list(tau)


def CGL2(N):
    theta = pi * (arange(0, N, dtype="float128")) / (N - 1)
    tau = cos(theta)[::-1]
    tau = (tau -tau[::-1])/2
    tau = array(tau, dtype="float64")
    return list(tau)
```
